## Maps

```python
def add (x,y):
    return x+y
L1=[1,2,3]
L2=[4,5,6]
result = map(add,L1,L2)
for i in result :
    print(i)
    
    
```

```python
def even (x):
    return x%2==0
L1 =[1,2,3]
L2 = [4,5,6]
result=map(even,L1)
for i in result:
    if (i%2!=0):
        print("False")
    else:
        print("True")    
    
```
