## For Loops



```python
List1 = [1,2,3,4,5,6,7]
for i in List1 :
    print(i)
```

```python
for i in range (11):
    print(i)
```

```python
for i in range (0,21):
    if i % 2==0 :
        print(i)
```

```python
numbers = [3,5,7,9,11,13]
sum = 0
for i in numbers :
    sum += i
print(sum)    
```

```python
for i in range (1,6):
    for j in range (1,6):
        print(f"{i}*{j} = ",i*j)
    print()   
```

```python
N =int(input("Enter your number : "))
sum = 0
for i in range (1,N+1):
    sum+=i
print("The sum of numbers is = ",sum)    
```

```python
N = int(input("Enter your number : "))
result =1
for i in range (1,N+1):
    result *=i
print(f"The facorial of {N} is = ",result)    
```
