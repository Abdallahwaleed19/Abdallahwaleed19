## Mean 

```python
import statistics
import matplotlib.pyplot as plt 
import scipy.stats as stats
X = list (map (int,input("Enter the values of x : ").split()))  # Values of X
Y = list (map (int,input("Enter the values of y : ").split())) # Values of Y 
# To Check before we start the mession 
if len(X) != len(Y):
       print("Error: The number of elements in X and Y must be the same.")
elif len(X) < 2:
       print("Error: X and Y must have at least two elements for correlation and regression.")
else :
# Mean
    mean_of_x = statistics.mean(X)
    mean_of_y = statistics.mean(Y)
    print(" Mean of X = " , mean_of_x) # Mean of X 
    print(" Mean of Y = " , mean_of_y) # Mean of Y
```

## Correlation Coefficient 

```python
correlation_of_x_y , p_value = stats.pearsonr(X, Y)
print("Correlation Coefficient between X and Y =", correlation_of_x_y) # Correlation Result
```

## linear Regression

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
print("\nLinear Regression Results:")
print("Slope (m):", slope)
print("Intercept (b):", intercept)
print("R-squared (r^2):", r_value ** 2)
print("P-value:", p_value)
print("Standard Error:", std_err)
```

## Generate the regression line

```python
regression_line = [slope * x + intercept for x in X]
```

## Estimation of variance of residuals

```python
n = len(X)
residuals = [(y - (slope * x + intercept)) for x, y in zip(X, Y)]
residual_variance = sum([r**2 for r in residuals]) / (n - 2)
print("\nResidual Variance (σ²):", residual_variance)
```

## Confidence interval for the slope (B)

```python
t_critical = stats.t.ppf(1 - 0.025, df=n - 2)  # Two-tailed 95% confidence
slope_conf_interval = (slope - t_critical * std_err, slope + t_critical * std_err)
print("\n95% Confidence Interval for the Slope (B):", slope_conf_interval)
```

## Hypothesis Testing for B (H0: B = 0)

```python
t_statistic = slope / std_err
p_value_hypothesis = 2 * (1 - stats.t.cdf(abs(t_statistic), df=n - 2))  # Two-tailed test
print("\nHypothesis Test for Slope (H0: B = 0):")
print("T-statistic:", t_statistic)
print("P-value:", p_value_hypothesis)
if p_value_hypothesis < 0.05:
        print("Conclusion: Reject the null hypothesis (H0). Slope is statistically significant.")
else:
        print("Conclusion: Fail to reject the null hypothesis (H0). Slope is not statistically significant.")
```

Mean
The mean, also known as the average, is a measure of central tendency. It is calculated by 
adding up all the values in a dataset and then dividing by the number of values. 
Formula:
Mean = (Sum of all values) / (Number of values) 
Pearson Correlation Coefficient:
The Pearson correlation coefficient (r) is a measure of the linear correlation between two 
variables. It ranges from -1 to +1. 
• r = -1: Perfect negative linear correlation • r = 0: No linear correlation 
• r = +1: Perfect positive linear correlation Formula:
r = (Σ(x_i - x̄)(y_i - ȳ)) / (√(Σ(x_i - x̄)^2) * √(Σ(y_i - ȳ)^2)) 
Multiple Linear Regression: 
The equation for a multiple linear regression model is: y = 
b0 + b1x1 + b2x2 + ... + bnxn where: 
• y is the dependent variable 
• x1, x2, ..., xn are the independent variables 
• b0, b1, b2, ..., bn are the coefficients Rules for Linear Regression:
1. Linearity: The relationship between the dependent and independent variables should be 
linear. 
2. Independence of Errors: The errors should be independent of each other. 
3. Homoscedasticity: The variance of the errors should be constant across all values of the 
independent variable. 
4. Normality of Errors: The errors should be normally distributed. 
By understanding these rules and formulas, you can effectively analyze and interpret data using 
mean, correlation, and linear regression.
