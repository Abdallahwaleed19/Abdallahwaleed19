## Dictionaries

```python
dict1 = {
    'A' : 1 ,
    'B' : 2 ,
    'C' : 3 ,
    'D' : 4 ,
    'E' : 5 ,
    }
dict1['E'] = 10
print(dict1)

```

```python
for key , value in dict1.items():
    print(f" key : {key} , Value : {value}")
```

```python
Abdallah ={
    "grade" : 'A' ,
    "age"   :  20 ,
}
Basmala = {
    "grade" : 'B' ,
    "age"   : 19 ,
}
print("Abdallah :", Abdallah)
print("Basmala  :",Basmala)
```

```python
scors_dict = {
    "Abdallah" : [100,90,85] ,
    "Sami"     : [80,60,89]  ,
    "Basmala"  : [90,93,80]  ,
}
average1 = sum (scors_dict["Abdallah"])/3 
average2 = sum (scors_dict["Sami"])/3 
average3 = sum (scors_dict["Basmala"])/3 
print(scors_dict)
print("The average of abdallah scors = ",average1)
print("The average of sami scors = ",average2)
print("The average of Basmala scors = ",average3)
```
