## Functions 


```python
def add (x,y):
    return x+y
add(1,2)
```

```python
def Multi (x,y):
    return x*y
Multi(3,4)
```

```python
def multi (*x):
    M=1
    for i in range (0,len(x)):
        M = M*x[i]
    return(M)
multi(1,2,3,4)    
```

```python
def my_function (fname):
    print(fname+"Refsnes")
my_function("Emil")    
```

```python
def Fahrenheit(celsius):
    return((celsius*9/5)+32)
Fahrenheit(37)
```

```python
def factorial (n):
    result = 1
    for i in range (1,n+1):
         result *=i
    return result     
n = int(input("enter your number : "))
factorial(n)    
```

```python
def prime_number (x):
    if x <2 :
        return False
    for i in range (2,x):
        if x%i==0:
            return False
    return True
prime_number(2)    
```

## Arbitrary Arguments (*args)

```python
def my_function (*kids):
    print("The youngest child is "+ kids[2])
my_function("Emil","Tobias","Linus")    
```

## Keywords Arguments

```python
def my_function (child1,child2,child3):
    print(" The youngest child is " + child3)
my_function(child1="Emil",child2="Tobias",child3="Linus")    
```

## Arbitrary Keywords Arguments (**kwargs)

```python
def my_function(**kid):
    print("His last name is "+ kid["Lname"])
my_function(Fname="Tobias",Lname="Refsnes")    
```

## Default Parameter Value 

```python
def my_function(country="Egypt"):
    print("I'm from "+country)
my_function()
my_function("France")    
```

## Return Values

```python
def my_function(X):
    return 5 * X
print(my_function(6))
```

## Lambda Functions


```python
Multiplication = lambda a,b : a*b
print(Multiplication(5,6))
```

```python
Result = lambda a,b,c : a**2 + b + c
print(Result(5,6,7))
```
